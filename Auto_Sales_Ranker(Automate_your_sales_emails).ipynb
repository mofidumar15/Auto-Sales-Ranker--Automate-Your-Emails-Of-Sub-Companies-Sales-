STEP 1: Generate Random Company Dataset

```python
import pandas as pd
import random

def generate_companies(n=250):
    companies = []

    for i in range(n):
        company = {
            "Company_Name": f"Company_{i+1}",
            "Total_Sales": round(random.uniform(10000, 1000000), 2),
            "Email": f"company{i+1}@example.com"
        }
        companies.append(company)

    return pd.DataFrame(companies)

df = generate_companies(250)
df.head()
```

STEP 2: Rank Companies by Sales

```python
def rank_companies(df):
    df_sorted = df.sort_values(by="Total_Sales", ascending=False)
    df_sorted["Rank"] = range(1, len(df_sorted) + 1)
    return df_sorted

ranked_df = rank_companies(df)
ranked_df.head()
```

STEP 3: Create Priority List (Top 50)

```python
def get_priority_list(df, top_n=50):
    return df.head(top_n)

priority_list = get_priority_list(ranked_df, 50)
priority_list
```

STEP 4: Save Priority List
Save as CSV

```python
priority_list.to_csv("priority_list.csv", index=False)
```

STEP 5: Automatic Email Sender Model

Please Note That:

Use Gmail App Password, NOT your real password
Enable "Less secure apps" or App Passwords

```python
import smtplib

def send_emails(priority_df, subject, message_body, sender_email, app_password):

    server = smtplib.SMTP("smtp.gmail.com", 587)
    server.starttls()
    server.login(sender_email, app_password)

    for _, row in priority_df.iterrows():
        receiver = row["Email"]
        company = row["Company_Name"]

        message = f"""
        Subject: {subject}

        Dear {company},

        {message_body}

        Regards,
        CEO
        """

        server.sendmail(sender_email, receiver, message)

    server.quit()
```

```python
subject = input("Enter Email Subject: ")
message_body = input("Enter Email Message: ")
sender_email = input("Enter Your Email: ")
app_password = input("Enter App Password: ")

send_emails(priority_list, subject, message_body, sender_email, app_password)
```
